# Mask R-CNN - Train on Shapes Dataset


This notebook shows how to train Mask R-CNN on your own dataset. To keep things simple we use a synthetic dataset of shapes (squares, triangles, and circles) which enables fast training. You'd still need a GPU, though, because the network backbone is a Resnet101, which would be too slow to train on a CPU. On a GPU, you can start to get okay-ish results in a few minutes, and good results in less than an hour.

The code of the *Shapes* dataset is included below. It generates images on the fly, so it doesn't require downloading any data. And it can generate images of any size, so we pick a small image size to train faster.

## Configurations

```python
class ShapesConfig(Config):
    """Configuration for training on the toy shapes dataset.
    Derives from the base Config class and overrides values specific
    to the toy shapes dataset.
    """
    # Give the configuration a recognizable name
    NAME = "shapes"

    # Train on 1 GPU and 8 images per GPU. We can put multiple images on each
    # GPU because the images are small. Batch size is 8 (GPUs * images/GPU).
    GPU_COUNT = 1
    IMAGES_PER_GPU = 8

    # Number of classes (including background)
    NUM_CLASSES = 1 + 3  # background + 3 shapes

    # Use small images for faster training. Set the limits of the small side
    # the large side, and that determines the image shape.
    IMAGE_MIN_DIM = 128
    IMAGE_MAX_DIM = 128

    # Use smaller anchors because our image and objects are small
    RPN_ANCHOR_SCALES = (8, 16, 32, 64, 128)  # anchor side in pixels

    # Reduce training ROIs per image because the images are small and have
    # few objects. Aim to allow ROI sampling to pick 33% positive ROIs.
    TRAIN_ROIS_PER_IMAGE = 32

    # Use a small epoch since the data is simple
    STEPS_PER_EPOCH = 100

    # use small validation steps since the epoch is small
    VALIDATION_STEPS = 5
    
config = ShapesConfig()
config.display()
```

```python
class CholoDataset(utils.Dataset):
    """Generates the shapes synthetic dataset. The dataset consists of simple
    shapes (triangles, squares, circles) placed randomly on a blank surface.
    The images are generated on the fly. No file access required.
    """
#     def __init__(self):
#         # Open file and extract data
#         f = open('./../bbox/server/log.json', 'r');
#         contents = f.readlines()
#         f.close()

#         # Convert the data into json format for easy extraction
#         self.json_format=json.loads(contents[0])        

    def load_shapes(self, vals):
        """Generate the requested number of synthetic images.
        count: number of images to generate.
        height, width: the size of the generated images.
        """
        # Add classes
        self.add_class("tools", 1, "grasper")
        self.add_class("tools", 2, "bipolar")
        self.add_class("tools", 3, "hook")
        self.add_class("tools", 4, "scissors")
        self.add_class("tools", 5, "clipper")
        self.add_class("tools", 6, "irrigator")
        self.add_class("tools", 7, "specimen bag")

        # Open file and extract data
        f = open('./../bbox/server/log.json', 'r');
        contents = f.readlines()
        f.close()

        # Convert the data into json format for easy extraction
        self.json_format=json.loads(contents[0]) 
        
        # Add images
        # Generate random specifications of images (i.e. color and
        # list of shapes sizes and locations). This is more compact than
        # actual images. Images are generated on the fly in load_image().
        for i in vals:
            #bg_color, shapes = self.random_image(height, width)
            img_src = './../bbox/src/img_vid1/'+self.json_format[i]['pic'][2:]
            img = imageio.imread(img_src)
            width = img.shape[1]
            height = img.shape[0]
            self.add_image("tools", image_id=i, path=img_src,
                           width=width, height=height,
                           bg_color=None, shapes=None)
        self._image_ids = vals

    def load_image(self, image_id):
        """Generate an image from the specs of the given image ID.
        Typically this function loads the image from a file, but
        in this case it generates the image on the fly from the
        specs in image_info.
        """
        self.image = imageio.imread('./../bbox/src/img_vid1/'+self.json_format[image_id]['pic'])

        return self.image

    def load_mask(self, image_id):
        """Generate instance masks for shapes of the given image ID.
        """                
        # Extract points data for each mask in specific image
        pts_all = []
        class_ids = []
        mask_arr = []
        for j in range(len(self.json_format[image_id]['mask'])):
            pts = []
            if (len(self.json_format[image_id]['mask'][j])!=0):
                for i in range(len(self.json_format[image_id]['mask'][j])):
                    x = self.json_format[image_id]['mask'][j][i]['x']
                    y = self.json_format[image_id]['mask'][j][i]['y']
                    pts.append([x,y])
                pts_all.append(pts)

        # Extract image information
        hgt = self.image.shape[0]
        wdt = self.image.shape[1]

        # Account for multiple masks
        # Each mask has a bounding box associated
        # with it and a class
        for m in range(len(pts_all)):
            mask = np.zeros((hgt,wdt))
            # Create the mask using OpenCV
            pts = np.array(pts_all)
            cv2.fillConvexPoly(mask, np.array(pts[m]), 1)
            mask = mask.astype(np.bool)
            mask_arr.append(mask) # Each mask is saved separately
            class_ids.append(self.json_format[image_id]['class'][m])
            #plt.figure()
            #plt.imshow(mask)

        mask_arr = np.array(mask_arr, dtype=np.bool)
        #mask_arr = np.reshape(mask_arr, [mask_arr.shape[1], mask_arr.shape[2], mask_arr.shape[0]])

        # Convert the arrays into forms that can be manipulated 
        # by the visualization tool
        if (mask_arr.shape[0]!=0):
            mask_arr = mask_arr.swapaxes(0,1).swapaxes(1,2)
            #print('~~~~~~~~~~~~~~~~~correct')
            #print(mask_arr.shape)
        else:
            mask_temp = np.zeros((hgt, wdt, 0))
            mask_arr = np.array(mask_temp, dtype=np.bool)
            #print(mask_arr.shape)
            
        class_ids = np.array(class_ids, dtype=np.int32)
        return mask_arr, class_ids
```

2. Fine-tune all layers. For this simple example it's not necessary, but we're including it to show the process. Simply pass `layers="all` to train all layers.

```python
# Train the head branches
# Passing layers="heads" freezes all layers except the head
# layers. You can also pass a regular expression to select
# which layers to train by name pattern.
model.train(dataset_train, dataset_val, 
            learning_rate=config.LEARNING_RATE, 
            epochs=1, 
            layers='heads')
```
